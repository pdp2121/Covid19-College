```python
# preprocess the tweets
# pip install ekphrasis

import re
from ekphrasis.classes.preprocessor import TextPreProcessor
from ekphrasis.classes.tokenizer import SocialTokenizer
from ekphrasis.classes.segmenter import Segmenter
from ekphrasis.dicts.emoticons import emoticons
from ekphrasis.dicts.noslang.slangdict import slangdict

text_processor = TextPreProcessor(
    # terms that will be normalized
    normalize=['url', 'email', 'percent', 'money', 'phone', 'user', 'time', 'date'],
    # terms that will be annotated
    annotate=['hashtag', 'allcaps', 'elongated', 'repeated', 'emphasis', 'censored'],
    fix_html=True,  # fix HTML tokens
    
    # corpus from which the word statistics are going to be used
    # for word segmentation 
    segmenter="twitter_2018", 
    
    # corpus from which the word statistics are going to be used 
    # for spell correction
    corrector="twitter_2018", 
    
    unpack_hashtags=True,  # perform word segmentation on hashtags
    unpack_contractions=True,  # Unpack contractions (can't -> can not)
    spell_correct_elong=True,  # spell correction for elongated words
    spell_correction=True,
    
    # select a tokenizer. You can use SocialTokenizer, or pass your own
    # the tokenizer, should take as input a string and return a list of tokens
    tokenizer=SocialTokenizer(lowercase=True).tokenize,
    
    # list of dictionaries, for replacing tokens extracted from the text,
    # with other expressions. You can pass more than one dictionaries.
    dicts=[emoticons, slangdict]
    )

segmenter=Segmenter(corpus="twitter_2018")

# replace the hashtags, emails, mentions, etc.
def preprocess(text):
    text = text.lower()
    text = re.sub(",", " ", text)
    text = re.sub(":", "", text)
    text = re.sub("@ ", "@", text)
    text = re.sub("# ", "#", text)
    text = re.sub('\S*@\S*\s?', '', text)
    text = re.sub(r"\"", "", text)
    text = re.sub("\'", "", text)
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r'\w*pic.twitter.co\w*', '', text)
    text = re.sub(r'\w*twitter.co\w*', '', text)
    text = re.sub(r'\w*twitter.com\w*', '', text)
    text = re.sub(r"./\S+", "", text)
    text = re.sub(r"@ \S+", "", text)
    text = re.sub(r"#\S+", "", text)
    text = re.sub(r'\n+', " ", text)
    text = re.sub(r"<.*?>", "", text)
    text = re.sub(r"<.*?>", "", text)
    text = re.sub("co vid", "covid", text)
    text = re.sub(r"\ss\s", " 's ", text)
    text = re.sub(r"\sm\s", " 'm ", text)
    text = re.sub(r"\sll\s", " 'll ", text)
    text = re.sub(r"\st\s", " 't ", text)
    text = re.sub(r"\sd\s", " 'd ", text)
    text = re.sub(r"\svir\s", " virus ", text)
    text = re.sub("rt", "", text)
    if get_lang(text) == 'en':
        return text.strip()
    else:
        return None
```

```python
# get non organizational users

non_org = []
tweet_count = 0

with open('./covid_users_100000_non-org.jsonl', 'w') as f:
    with progressbar.ProgressBar(max_value=len(users)) as bar:
        for i in range(len(users)):
            if users[i]['id'] not in org:
                pred = predictions[users[i]['id']]
                users[i]['age'] = pred['age']
                users[i]['gender'] = pred['gender']
                users[i]['org'] = pred['org']
                users[i]['is_student'] = None
                tweet_file = './data/user-timelines/'+str(users[i]['id'])+'_tweets.jsonl'
                tweets = open(tweet_file, 'r').readlines()
                tweets = [json.loads(t)['text'] for t in tweets]
                
                processed_tweets = []
                for tweet in tweets:
                    tweet = tweet.encode('ascii', 'ignore').decode('utf-8')
#                     print(tweet)
                    tokens = text_processor.pre_process_doc(tweet)
                    tokens = [segmenter.segment(t) for t in tokens]
                    text = " ".join(tokens)
                    processed_tweets.append(preprocess(text))

                processed_tweets = list(filter(None, processed_tweets))
#                 processed_tweets = [t+'.' for t in processed_tweets if not t.endswith('.')]
                tweet_count += len(processed_tweets)
#                 processed_tweets = ' '.join(processed_tweets)

                users[i]['tweets'] = processed_tweets
                non_org.append(users[i])
                f.write(json.dumps(users[i])+'\n')
            bar.update(i)
```

```python
# stanford nlp speech tagger
# download the latest version and put in ./tools, then run "stanford_corenlp.sh"

from stanfordcorenlp import StanfordCoreNLP
import logging

class StanfordNLP:
    def __init__(self, host='http://localhost', port=9000):
        self.nlp = StanfordCoreNLP(host, port=port, timeout=30000)  # , quiet=False, logging_level=logging.DEBUG)
        self.props = {
            'annotators': 'tokenize,ssplit,pos,lemma,parse,depparse',
            'pipelineLanguage': 'en',
            'outputFormat': 'json'
        }

    def word_tokenize(self, sentence):
        return self.nlp.word_tokenize(sentence)

    def pos(self, sentence):
        return self.nlp.pos_tag(sentence)

    def ner(self, sentence):
        return self.nlp.ner(sentence)

    def parse(self, sentence):
        return self.nlp.parse(sentence)

    def dependency_parse(self, sentence):
        return self.nlp.dependency_parse(sentence)

    def annotate(self, sentence):
        return json.loads(self.nlp.annotate(sentence, properties=self.props))

    @staticmethod
    def tokens_to_dict(_tokens):
        tokens = defaultdict(dict)
        for token in _tokens:
            tokens[int(token['index'])] = {
                'word': token['word'],
                'lemma': token['lemma'],
                'pos': token['pos']
            }
        return tokens

nlp = StanfordNLP()
```

```python
# run the tagger and update the data file

non_org = open('./covid_users_100000_non-org.jsonl', 'r').readlines()
non_org = [json.loads(u) for u in non_org]

with progressbar.ProgressBar(max_value=len(non_org)) as bar:
    for i in range(len(non_org)):
        processed = []
        for tweet in non_org[i]['tweets']:
            t = {}
            t['text'] = tweet
            t['word'] = ''
            t['lemma'] = ''
            t['pos'] = ''
            for sent in nlp.annotate(tweet)['sentences']:
                for token in sent['tokens']:
                    t['word'] = " ".join([t['word'], token['word']])
                    t['lemma'] = " ".join([t['lemma'], token['lemma']])
                    t['pos'] = " ".join([t['pos'], token['pos']])
            processed.append(t)
            non_org[i]['tweets'] = processed
        bar.update(i)
```
